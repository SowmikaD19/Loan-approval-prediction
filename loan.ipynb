```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
```

```python
df=pd.read_csv('train.csv')
df
```

```python
df.dtypes
```

```python
df['loan_status'].value_counts()
```

```python
df['education'].value_counts()
```

```python
df['self_employed'].value_counts()
```

```python
# Clean the column by replacing non-numeric values with a default value, for example, -1
df['self_employed'] = df['self_employed'].replace({' No': 0, ' Yes': 1})
#df.dtypes
```

```python
df['education'] = df['education'].replace({' Graduate': 1, ' Not Graduate': 0})
```

```python
df['loan_status'] = df['loan_status'].replace({' Approved': 1, ' Rejected': 0})
```

```python
df['loan_status'].value_counts()
```

```python
df
```

```python
df.dtypes
```

```python
X=df.drop(["loan_id","loan_status"],axis=1)
y=df["loan_status"]
```

```python
#Train test split
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=0,test_size=0.25)
```

```python
#Model Trainig
model = RandomForestClassifier(n_estimators=100, random_state=0)
model.fit(X_train, y_train)
```

```python
#Printed the training score
print(f"Training Score:{model.score(X_train,y_train)*100}%")
#Printed the testing score
print(f"Test score:{model.score(X_test,y_test)*100}%")
```

```python
#testing
test = pd.read_csv('test.csv')
test
```

```python
test['self_employed'] = test['self_employed'].replace({' No': 0, ' Yes': 1})
test['education'] = test['education'].replace({' Graduate': 1, ' Not Graduate': 0})
test['loan_status'] = test['loan_status'].replace({' Approved': 1, ' Rejected': 0})
test
```

```python
test1=test.drop(["loan_id","loan_status"],axis=1)
test1
```

```python
pred = model.predict(test1)
pred
```

```python
test["Pred"]=pd.DataFrame(pred)
test
```

```python
# Calculate the number of correct predictions
correct_predictions = (test['loan_status'] == test['Pred']).sum()

# Calculate the total number of predictions
total_predictions = len(test)

# Calculate accuracy
accuracy = (correct_predictions / total_predictions) * 100

# Display the result
print(f"Accuracy: {accuracy:.2f}%")
```
